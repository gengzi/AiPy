```python
from transformers import AutoModelForCausalLM, AutoTokenizer
from transformers import AutoTokenizer

from awq.quantize.pre_quant import apply_awq, get_awq
from awq.quantize.quantizer import pseudo_quantize_model_weight



model_path =  f"/mnt/e/models/DeepSeek-R1-Distill-Qwen-1.5B";
quant_config = {"zero_point": True, "q_group_size": 128, "w_bit": 4, "version": "GEMM"}
quant_path = "/mnt/e/models/quantization/deepseek1.5_quantization"
newmode = AutoModelForCausalLM.from_pretrained(model_path, device_map="cuda:0");

tokenizer = AutoTokenizer.from_pretrained(model_path, trust_remote_code=True)

calibration_dataset = [
    "这是一个校准用的示例句子。",
    "另外一个用于校准的句子。"
    # 可以添加更多句子
]

# 配置AWQ参数
awq_config = {
    "w_bit": 4,  # 量化位宽，通常为4位
    "q_group_size": 128,  # 量化组大小
    "no_zero_point": False,  # 是否使用零点
    "use_cache": True  # 是否使用缓存
}
# 计算AWQ参数
awq = get_awq(model, tokenizer, calibration_dataset, **awq_config)
# 应用AWQ量化
apply_awq(model, awq)
# 伪量化模型权重
pseudo_quantize_model_weight(model, w_bit=awq_config["w_bit"], q_group_size=awq_config["q_group_size"])

newmode.save_quantized(quant_path)
tokenizer.save_pretrained(quant_path)
```
